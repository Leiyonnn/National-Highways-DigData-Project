```python
import pandas as pd

file_path = 'DigData Step Up.xlsx'
campaigns_data = pd.read_excel(file_path, sheet_name='Campaigns data')
study_data = pd.read_excel(file_path, sheet_name='Study Data')
campaigns_data.head()
merged_data = pd.merge(campaigns_data, study_data, on=['study_id', 'cell_id'], how='inner')
merged_data.head()
```

```python
success_by_region = merged_data['advertiser_region'].value_counts(normalize=True) * 100
success_by_region
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=success_by_region.index, y=success_by_region.values)
plt.title("Regional Success Distribution")
plt.xlabel("Region")
plt.ylabel("Success Rate (%)")
plt.show()
```
